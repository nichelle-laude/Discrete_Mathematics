# 1. Proving Identities

$\textbf{Given:}\, A \ and \ B\, are \ sets,\ \exists \ x \ such  \ that \ (x \in A \land x \in B).$


$ \textbf{To Show:} \ A \setminus B = A \cap B^c\$

$\begin{align*}A \setminus B &= \{ x \mid x \in A \text{ and } x \notin B \} \\&= \{ x \mid (x \in A \text{ and } x \in B^c) \} \\&= A \cap B^c \quad \text{(by definition of set difference)} \\&= A \cap B \quad \text{(since } x \in B \text{ implies } x \notin B^c)\end{align*}$


$\textbf{To Show:} \ B = (B \cap A) \cup (B \cap A^c)$

$\begin{align*}(B \cap A) \cup (B \cap A') &= \{ x \mid (x \in B \text{ and } x \in A) \text{ or } (x \in B \text{ and } x \notin A) \} \\&= \{ x \mid x \in B \} \quad \text{(since } x \in A \land x \in B \text{ implies } x \in B) \\&= B\end{align*}$



$\textbf{To Show:}\ B \setminus A = B \cap A^c$

$\textbf{Proof:}$

$\begin{align*}B \setminus A &= \{ x \mid x \in B \text{ and } x \notin A \} \\&= \{ x \mid (x \in B \text{ and } x \in A') \} \\&= B \cap A' \quad \text{(by definition of set difference)}\end{align*}$



$\textbf{To Show:} \ A \cup B = A \cup (B \cap A^c)$


$\textbf{Proof:}$

$\begin{align*}A \cup B &= \{ x \mid x \in A \text{ or } x \in B \} \\&= \{ x \mid (x \in A \text{ or } (x \in B \text{ and } x \notin A)) \} \\&= \{ x \mid (x \in A \text{ or } x \in B) \text{ and } (x \in A \text{ or } x \notin A) \} \\&= \{ x \mid x \in A \text{ or } x \in B \} \\&= A \cup (B \cap A') \quad \text{(since } x \in A \text{ or } x \notin A \text{ is always true)}\end{align*}$

$\textbf{To Show:}\left( \bigcup_{i=1}^{\infty} A_i^c \right)^c = \bigcap_{i=1}^{\infty} A_i$

$\textbf{Proof:}$

$$Let \ (x) be \ an \ element \ in \ \left( \bigcup_{i=1}^{\infty} A_i^c \right)^c\,which \ means (x) is \ in \ the \ complement \ of \ the \ union \ of \ all (A_i^c). This \ implies \ (x)  \ is \ not \ in \ the \ complement \ of \ at \ least \ one \ of \ the \ sets (A_1^c, A_2^c, \ldots). In \ other \ words, (x) is \ in \ at \ least \ one \ of \ the \ sets (A_1, A_2, \ldots), and \ thus:$$

$$x \in \bigcap_{i=1}^{\infty} A_i$$

$$Conversely, \ let \ (x)  \ be \ an \ element \ in \  (\bigcap_{i=1}^{\infty} A_i). This \ means \ (x) \ is \ in \ every \ set (A_i), \ which \ implies (x) is \ not \ in \ the \ complement\  of \ any \ (A_i^c).\  Hence, \ (x) \ is not \ in \ the \ union \ of \ any \ (A_i^c), \ and \ we \ have:$$

$$x \in \left( \bigcup_{i=1}^{\infty} A_i^c \right)^c$$

$$Thus, \ the \ sets \left( \bigcup_{i=1}^{\infty} A_i^c \right)^c) \ and \ (\bigcap_{i=1}^{\infty} A_i)  \ contain \ the \ same \ elements, \ and \ they \ are \ equal.$$


$\textbf{To Show:}|A \cup B| = |A| + |B| - |A \cap B|$

$\textbf{Proof:}$

$$Let (n(A) \ denote \ the \ number \ of \ elements \ in \ set (A). \ The \ union \ (A \cup B)  \ contains \ all \ elements \ from \ (A) \ and \ (B), \ but \ we \ have \ counted \ the \ common \ elements \ (A \cap B) \ twice. \ Therefore, \ the \ total \ number \ of \ distinct \ elements \ in \ 
 (A \cup B) \ can \ be \ calculated \ as \ the \ sum \ of \ the \ elements \  in \ (A) \ and \ (B), \ minus \ the \ number \ of \ common \ elements:$$

$$|A \cup B| = n(A) + n(B) - n(A \cap B) = |A| + |B| - |A \cap B|$$

```python
from matplotlib_venn import venn3, venn3_circles
import matplotlib.pyplot as plt

# Given data
newsweek_only = 65 - (20 + 25 - 8)
times_only = 45 - (20 + 15 - 8)
fortune_only = 42 - (25 + 15 - 8)
newsweek_and_times = 20 - 8
newsweek_and_fortune = 25 - 8
times_and_fortune = 15 - 8
all_three = 8

#Let X be the people who read at least one of the three magazines
X = 65 + 45 + 42 - 25 - 20 - 15 + 8

#Let Y be the total of people who have read exactly one magazine
fortune_readers = 42 - 25 - 15 + 8
newsweek_readers = 65 - 25 - 20 + 8
times_readers = 45 - 15 -20 + 8
Y = fortune_readers + newsweek_readers + times_readers

# Create the Venn diagram
venn = venn3(subsets=(newsweek_only, times_only, newsweek_and_times,
                      fortune_only, newsweek_and_fortune, times_and_fortune, all_three),
             set_labels=('Newsweek', 'Times', 'Fortune'))

plt.title("Venn Diagram of Magazines")
plt.show()

print("Number of people who read at least one of the 3 magazines:", X)
print("Total number of people who read at exactly one magazine:", Y)
```
